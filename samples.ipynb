```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
from datetime import time
```

```python
df = pd.read_csv('datasets/90460.csv', dtype=str)
df.columns
```

```python
df['collectdt'] = pd.to_datetime(df['Collect Date'], format = '%Y-%m-%d %H:%M:%S')
```

```python
df.head()
cfu = df[(df['PARAMID']=='CFU') & (df['SAMPLESTATUS']=='Approved')]
cfu[['DISPLAYVALUE']] = cfu[['DISPLAYVALUE']].astype(float)
x = cfu[cfu['DISPLAYVALUE'] > 0]
x
```

```python
onesp = cfu[cfu['SP Code'] == '71-114-D-AS.1']
img = onesp.reindex(columns=['DISPLAYVALUE', 'collectdt']).sort_values('collectdt')
img.set_index('collectdt', inplace=True)
img
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(img, 'ko--')
```
